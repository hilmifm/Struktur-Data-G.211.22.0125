```python
input_list = ['l', 'a', 'm', 'b', 'o', 'r', 'g', 'h', 'i', 'n', 'i']

# Menguraikan setiap karakter dan membangun kata "lamborghini"
output_string = ""
output_string += input_list[0]  # l
output_string += input_list[1]  # a
output_string += input_list[2]  # m
output_string += input_list[3]  # b
output_string += input_list[4]  # o
output_string += input_list[5]  # r
output_string += input_list[6]  # g
output_string += input_list[7]  # h
output_string += input_list[8]  # i
output_string += input_list[9]  # n
output_string += input_list[10]  # i

print(output_string)  # Output: "lamborghini"
```
